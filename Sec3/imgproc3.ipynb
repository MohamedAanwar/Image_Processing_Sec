## Sec3-Image Processing

```python
# import libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read image
img=cv2.imread("cameraman.jpg")
```

```python
img.shape
```

```python
#Convert to gray

gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray image',gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
gray_img.shape
```

```python
# Calc histogram

hist_img=cv2.calcHist([gray_img],[0],None,[256],[0,255])
```

```python
# Show histogram

plt.hist(gray_img.ravel(),256,[0,255])
plt.title('Histogram')
plt.show()
```

```python
# Calc histogram Numpy

hist,bins=np.histogram(gray_img.ravel(),bins=256,range=(0,255))
```

```python
# Show histogram

plt.hist(gray_img.ravel(),256,[0,255])
plt.title('Histogram')
plt.show()
```

```python
# Histogram Equalization

eq_img=cv2.equalizeHist(gray_img)
```

```python
# show to image in plot

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(gray_img,cmap='gray')

plt.subplot(1,2,2)
plt.title("equalized image")
plt.imshow(eq_img,cmap='gray')
```

```python
# Histrogram to original image

plt.hist(gray_img.ravel(),256,[0,255])
plt.title('Histogram-Original image')
plt.show()
```

```python
# Histogram to Equalized image

plt.hist(eq_img.ravel(),256,[0,255])
plt.title('Histogram-Equalized image')
plt.show()
```
